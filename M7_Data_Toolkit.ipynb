```python
# 1. Create a 2D NumPy array and calculate the sum of each row:
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
row_sums = np.sum(arr, axis=1)
print(row_sums)
```

```python
# 3. Create a line plot using Matplotlib:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y = [10, 20, 25, 30]

plt.plot(x, y)
plt.title("Simple Line Plot")
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
plt.show()
```

```python
# 4. Create a histogram using Seaborn:
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 20, 30, 40, 40, 50]
sns.histplot(data, bins=5)
plt.title("Histogram")
plt.show()
```

```python
#5. Create a heatmap using Seaborn and visualize a correlation matrix:
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({
    'A': [1, 2, 3, 4],
    'B': [4, 3, 2, 1],
    'C': [10, 20, 30, 40]
})
corr = df.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
# 6. Create a bar plot using Matplotlib:
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C']
values = [10, 20, 15]

plt.bar(categories, values)
plt.title("Bar Plot")
plt.show()
```

```python
#7. Create a scatter plot using Seaborn:
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame({'x': [1, 2, 3], 'y': [4, 5, 6]})
sns.scatterplot(x='x', y='y', data=df)
plt.title("Scatter Plot")
plt.show()
```

```python
#8. Create a box plot using Seaborn:
import seaborn as sns
import matplotlib.pyplot as plt

data = [10, 20, 30, 40, 50, 60, 70]
sns.boxplot(data=data)
plt.title("Box Plot")
plt.show()
```

```python
#9. Create a line plot with multiple lines using Matplotlib:
import matplotlib.pyplot as plt

x = [1, 2, 3, 4]
y1 = [10, 20, 25, 30]
y2 = [15, 18, 22, 27]

plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.legend()
plt.title("Multiple Line Plot")
plt.show()
```

```python
#10. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold:
import pandas as pd

df = pd.DataFrame({'Name': ['Alice', 'Bob'], 'Score': [85, 60]})
filtered_df = df[df['Score'] > 70]
print(filtered_df)
```

```python
#11. Create a pivot table in Pandas:
import pandas as pd

df = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Alice', 'Bob'],
    'Subject': ['Math', 'Math', 'Science', 'Science'],
    'Score': [85, 90, 88, 92]
})
pivot = df.pivot_table(values='Score', index='Name', columns='Subject')
print(pivot)
```
