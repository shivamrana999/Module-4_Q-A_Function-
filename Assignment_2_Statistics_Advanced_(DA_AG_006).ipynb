```python
# Question 1: What is a random variable in probability theory?
- A variable whose values depend on outcomes of a random phenomenon.
Example: Rolling a dice → random variable X can take values {1,2,3,4,5,6}.
```

```python
# Question 2: What are the types of random variables?
- Discrete: Countable outcomes (dice rolls, number of students).
- Continuous: Infinite possible values (height, weight, time).
```

```python
# Q3: Discrete vs Continuous distributions
- Discrete distribution: Probabilities assigned to specific values. Example: Binomial distribution.
- Continuous distribution: Probabilities over intervals. Example: Normal distribution.
```

```python
# Question 4: What is a binomial distribution, and how is it used in probability?
- - Models number of successes in fixed trials with probability p.
Example: Probability of getting 3 heads in 5 coin tosses
```

```python
# Question 5: What is the standard normal distribution, and why is it important?
- Mean = 0, Standard deviation = 1.
- Important because many statistical tests rely on it (z-scores, hypothesis testing).
```

```python
# Question 6: What is the Central Limit Theorem (CLT), and why is it critical in statistics?
- States that sample means from large samples approximate a normal distribution, regardless of population distribution.
- Critical for confidence intervals and hypothesis testing.
```

```python
# Question 7: What is the significance of confidence intervals in statistical analysis?
- Range of values likely to contain the true population parameter.
- Example: 95% CI for mean income = [45k, 55k]
```

```python
# Question 8: What is the concept of expected value in a probability distribution?
# - Weighted average of all possible outcomes.
#Example: Rolling a dice →
Expected_value = (1+2+3+4+5+6)/6
Print=(Expected_value)
```

```python
# Question 9: Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.
import numpy as np
import matplotlib.pyplot as plt

data = np.random.normal(50, 5, 1000)

print("Mean:", np.mean(data))
print("Std Dev:", np.std(data))

plt.hist(data, bins=30, edgecolor='black')
plt.title("Normal Distribution (mean=50, std=5)")
plt.show()
```

```python
# Question 10: You are working as a data analyst for a retail company. The company has collected daily sales data for 2 years and wants you to identify the overall sales trend.
import scipy.stats as st
import numpy as np

daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

mean_sales = np.mean(daily_sales)
std_err = st.sem(daily_sales)
confidence_interval = st.t.interval(0.95, len(daily_sales)-1, loc=mean_sales, scale=std_err)

print("Mean Sales:", mean_sales)
print("95% Confidence Interval:", confidence_interval)
```
